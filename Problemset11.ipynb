```python
A* genau wie Dijkstra Algoritmus eignet sich sehr gut für die Klausur! 
Allerdings mit kleinerem Graph!


Wdh Dijkstra: Findet Kürzeste Strecke von Ausgangsknoten zu allen anderen Knoten!
#10.1. Dijkstra-Algorithm, Numerische Mathematik (1959)
Unseen / Fringe="Open-List" / Tree="Closed List" Knoten

- berechnet den kürzesten Weg von einem Startknoten zu allen anderen Knoten (optimal nicht nur Näherung)
- zum Anfang verwenden wir die Kosten der direkten Pfade
- in jedem Schritt überprüfen wir für alle, ob man sie verbessern kann, indem man Zwischenknoten sucht, die den Pfad 
verkürzen
- am Anfang werden alle Distanzen zu "unendlich" gesetzt
- Am Anfang enthält unsere Menge, S, der bereits abgearbeiten Knoten, nur den Wurzelknoten/Root
- in jedem Schritt fügen wir nun den Knoten mit der kürzsten Distanz hinzu und updaten die Tabelle mit den Distanzen



Wdh A*: Findet kürzeste Strecke von Ausgangsknoten zu Zielknoten (Hier: Von Heidelberg nach Berlin)
    ~zu Dijkstra, nur dass er im Gegensatz zu Dijkstra nicht nur "zurück" 
    sondern zusätzlich auch "nach vorne" schaut!
    
    -in jedem Schritt wird nun deshalb der Knoten "entwickelt" also der "Closed List" hinzu gefügt,
    der den niedristen Wert "f" hat! Wobei "f" gleich der Summe von "g" und "n" ist!
    
    -f(n)=g(n)+h(n)
    wobei g(n) wie bei Dijkstra die Kosten von Heidelberg bis zum jeweiligen Knoten ausgibt!
    während h(n) eine admissible/zulässige Heuristik für die Entfernung von Heidelberg nach Berlin ist
```
