```python
def function0(a):
    if a==0:
        return True
    if a==1:
        return False
    if (a<0):
        return function0(-a)
    return function0(a-2)
```

```python
O(1): return true/false, return f0(-a) (kommt jeweils genau einmal vor) und O(1)+O(1)+O(1)=O(1)

O(n): return f0(a-2); wird a/2 mal aufgerufen

keine andere Laufzeiten.

Formel: O(3*1) + O(n) = O(n)    
```

```python
def function2(a, n):
    x=1.0
    for i in range(n):
        x=0.5*(x+ (a/x))
    return x
```

```python
Wdh BinaryTree /  Bin√§rbaume
Hierarchische Datenstruktur (ungleich: Linked Lists, Arrays...)
bez√ºglich Zugriffszeit/Suche: array>BinaryTrees>Linked List
bez√ºglich Insertion/Deletion: Linked List > Binary Trees > Array
    
Beispiel: Suchbaum aus geordenetem Array erstellen: [1,2,5,8,23]
```

```python
If you want to read more about Binary (Search) Trees before Chapter4:
    https://www.cs.cmu.edu/~adamchik/15-121/lectures/Trees/trees.html
```

```python
#Pseudocode Find-Value-In-Binary-Search-Tree
def f9(value, t):
    t_current = t.root
    while(t_current Is Not Leaf):
        if value == t_current.value:
            return True
        if value > t_current.value:
            t_current = t_current.right
        else:
            t_current = t_current.left
    return value == t_current.value
```

```python
- Anzahl Vergleiche entspricht maximal der maximalen Pfadtiefe des Baumes

- Sei ‚Ñé die H√∂he des Baumes, dann ist die Komplexit√§t der Suche ùëÇ(‚Ñé)
    wegen 
    
- h < n-2
- h > log2(n+1)-1

So, if the tree has "n" nodes: The number of comparison is maximal:
    Best case O(log2(n)) and Worst Case (0(n))
    
O(log(n)): For a complete binary tree (A binary tree, which is completely filled, with the possible exception of the bottom level, which is filled from left to right)
    
O(1) Wertdefinierung und return
O(log(n)) f√ºr Anzahl an Vergleichen, wenn n ist die anzahl an Knoten im kompletten Bin√§rbaum
Formel: O(1) + O(log2(n)) = O(log2(n))    
```

```python
#java search same elements in a 
int f1A(char[] a, char[] b){
    int count = 0;
    for(char c: a){
        for(char d: b){
            if(c == d) count++;
        }
    }
    return count;
}
```

```python
O(1) Wertdefinerung und return
O(n*m) F√ºr Verschachtelter Loop, wobei arrays "n" bzw "m" lang sind
Formel O(1)+O(n*m)=O(n*m)
```
