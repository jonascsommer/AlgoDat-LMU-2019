```python
WIEDERHOLUNG Sortieren:
    
    -Elemente eines Array neu arrangieren. Jedes Item enthält einen Key nach dem die Items geordnet werden sollen.
    
    -Immer erster Schritt um Daten zu nutzen
    
    -Früher 30% aller CPU time für Sortieren
    -Heute weniger: Nicht weil weniger Sortieren, sondern weil relativ effiziente Algorithmen vorliegen
    
    -Warum nicht einfach System-Algorithmen verwenden?:
        + Gutes Anwendungsbeispiel für Analyse von Algorithmen
        + Sortieralgorithmen sind oft Ausgangspunkt zur Lösung anderer Probleme
        
    -Wir beurteilen Algorithmen nach Worst-Case/Best-Case/Average-Case von:
        +Anzahl von Vergleichen
        +Anzahl von Vertauschungen
        (+Stabilität)
```

```python
def BubbleSort(arr):
    v = 0   #Anzahl Vergleiche
    s = 0   #Anzahl swaps
    for i in range(len(arr)-1,0,-1): #from n-1 to 1
        for j in range(0,i):
            v = v + 1
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                s = s + 1
            print(arr, arr[j], arr[j+1], "Vergleich:", v, "Swap:", s)
```

```python
INSERTION SORT
```

```python
-nächsteinfacher Sortier-Algorithmus:

-Vom Prinzip sehr ähnlich wie BubbleSort:
    AUCH immer vertauschen/swappen von benachbarten Elementen
    NUR jetzt von links nach rechts durchgehen und immer alle Elemente 
        links des aktuellen Pivot-Elementes sortieren.
```

```python
def InsertionSort(arr):
    v = 0   #Anzahl Vergleiche
    s = 0   #Anzahl swaps
    for i in range(1,len(arr)):
        for j in range(i,0,-1):
            v = v + 1
            if arr[j] < arr[j-1]:
                arr[j],arr[j-1] = arr[j-1],arr[j]
                s = s + 1
            print(arr, arr[j-1], arr[j], "Vergleich:", v, "Swap:", s)
            
```

```python
a=[6,5,4,3,2,1]
b=[1,2,3,4,5,6]
InsertionSort(a)
```

```python
SELECTION SORT
```

```python
- Anfangen beim ersten item
- Dann durchgehen und das kleinste aller Elemente raussuchen
- Dieses Element mit dem ersten vertauschen

=>Effekt: Nicht so viele Vertauschungen notwendig
```

```python
def SelectionSort(arr):
    v = 0   #Anzahl Vergleiche
    s = 0   #Anzahl swaps
    for i in range(len(arr)-1):
        min = i
        for j in range(i+1,len(arr)):
            v = v + 1
            print(arr, "Vergleich:", v, "Swap:", s, "Vergleich von:", arr[j], arr[min] )
            if arr[j] < arr[min]:
                min = j
        if i != min:
            arr[i], arr[min] = arr[min], arr[i]
            s = s+1
            print("Swap of:" , arr[i], arr[min])
```

```python
a=[6,5,4,3,2,1]
b=[1,2,3,4,5,6]
SelectionSort(a)
```

```python
SelectionSort(b)
```

```python
Anzahl Vergleiche für Selection Sort:
    
    
    In jedem der 𝑛 Durchläufe 𝑛 − 𝑖 Vergleiche --> n*(n-1)/2 --> O(n^2)
    (unabhängig von der Länge: Denn SelectionSort ist kein adaptiver Sortieralgorithmus.
     Es tut also nichts zur Sache, wie ein Array ausschaut; man benotigt immer die selbe 
     Anzahl an Vergleichsoperationen.)
            
            
            
    
    
Anzahl Vertauschungen für Selection Sort:
    
    
    
    Anzahl der Vertauschungen, ist linear in n --> 0(n)
    
    Das ist der Vorteil vom Selection Sort, was diesen Sortier-Algorithmus auszeichnet

    Auch hier: maximale Zahl von Swaps=(n-1)
        Worst Case Beispiel: d=[6, 5, 4, 2, 1, 3 ]
```

```python
d=[6, 5, 4, 2, 1, 3 ]
SelectionSort(d)
```

```python
ZUSAMMENFASSUNG BUBBLE SORT

-> BAD: Anzahl der Vergleiche unabhängig von der Vorsortierung

-> BAD: Laufzeit: Anzahl an Vergleichen: Worst Case / Average Case: O(n^2)

-> GOOD: Bubble Sort ist stabil
    
-> GOOD: Trivialstes Beispiel für einen Sortieralgorithmus (Deshalb wurde er überhaupt hier behandelt)
```

```python
ZUSAMMENFASSUNG INSERTION SORT

--> BAD: Laufzeit: Anzahl an Vergleichen: Worst Case / Average Case: O(n^2)

--> GOOD: Stable!
    
--> GOOD: Kann zum "finishen" von bereits ziemlich gut vorsortierten Arrays verwendet werden (TimSort, ShellShort)
```

```python
ZUSAMMENFASSUNG SELECTION SORT

--> BAD: Selection Sort profitiert nicht von Vorsortierung
            VERY STUPID: -> O(n^2) sogar für Best Case: komplett vorsortierten Array
    
-> BAD: Laufzeit: Anzahl an Vergleichen: Worst Case / Average Case: O(n^2)
    
--> BAD: Unstable! 
    Also bleibt die relative Ordnung von items mit dem gleichen Key-Value nicht notwendigerweise
        erhalten!
        Bsp: [3,2,3,1]->First Iteration->[1,2,3,3]
            Reihenfolge der "3"s hat sich vertauscht

--> GOOD: Selection Sort hat minimale Datenbewegung / Minimale Anzahl an Swaps! 
            (Kein Anderer Sort-Algorithmus hat diese Eigenschaft)
            (Also wenn Swaps teuer (Bsp. Container) dann gut)
```
