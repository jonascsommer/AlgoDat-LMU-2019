```python
(i) search(20, 42)
(ii) search(20, 1)
(iii) search(20, 35)
(iv) search(20, 20)
(v) search(20, 23)
```

```python
Suche im BST

Algorithmus:
def search(self, key):
    if key == self.key:
        return self.value
    elif key < self.key and self.left != None:
        return self.left.search(key)
    elif key > self.key and self.right != None:
        return self.right.search(key)
    

In Worten:
Sei v der aktuelle Knoten:
    - Falls schlüssel/key==value(v) --> gefunden
    - Sonst:
        -falls v Blatt ist --> key ist nicht enthalten
        -falls key<value(v) --> suche im linken Teilbaum
        -falls key>value(v) --> suche im rechten Teilbaum
```

```python
(i) get(42) = 42 (20 → 42)
(ii) get(1) = 1 (20 → 16 → 5 → 1)
(iii) get(35) = None (20 → 42 → 32)
(iv) get(20) = 20 (20)
(v) get(23) = None (20 → 42 → 32 → 35)
```

```python
Einfügen in BST:
    "insert(key,value)"
    
    -suche key im Baum
    -wenn key bereits existiert, ersetze vorheriges Objekt
    -Falls nicht, endet die Binäresuche in einem Blatt oder Innerem Knoten mit nur einem Nachfolger
    -Füge neuen Knoten (key,value) hinzu, als linken oder rechten Folgeknoten
```

```python
Löschen aus BST:
    "remove(key,value)"
    
    -suche key im Baum 
    -->falls key nicht existiert passiert nichts
    -->sonst fallunterscheiden:
        ->1. Falls v ein Blatt ist, lösche Zeiger auf das Blatt
        ->2. Falls v nur ein Kind hat, lösche Zeiger auf v und verbinde Vaterknoten mit Kindknoten von v
        ->3. Falls v zwei Kinder hat, finde rechtesten Knoten im linken Teilbaum von v und tasche die Knoten.
                Lösche Zeiger auf v.
```

```python
AVL-Baum: BST mit AVL Kriterium.
    AVL-Kritierum: Balance b muss {-1,0,1} sein für alle Knoten
        wobei Balance b:=Höhe(rechter Teilbaum)-Höhe(linker Teilbaum)
            
```

```python
Warum wollen wir überhaupt balancierte Bäume?




-damit wir logaritmisch schnell in n suchen können!
-AVL Baum ist maximal 44% höher ein maximal ausgeglichener Baum!
```

```python
EINFACH-ROTATION
```

```python
    ->Einfach-Rotation brauchen wir if and only if 
    |balance(Kritischer Knoten)-balance(Knoten Drunter in Krit. Richtung)|=1
```

```python
Ok, was machen wir wenn durch Einfügen oder Löschen unser Baum nicht mehr balanciert ist?




->wieder balancieren durch Rotation!
```

```python
Schritte in einfacher Rotation:
    
1.ein Teilbaum ist zu "schwer"/"zu tief" -> wir "ziehen so lange am kürzeren Ende"/
"mit dem Kritischen Punkt als Aufhänger, bis wir Balance hergestellt haben"
2. Dann hängen wir den mittleren Teilbaum dahin "wo's passt"
```
