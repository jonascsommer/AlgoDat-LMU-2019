```python
def SelectionSort(arr):
    v = 0   #Anzahl Vergleiche
    s = 0   #Anzahl swaps
    for i in range(len(arr)-1):
        min = i
        for j in range(i+1,len(arr)):
            v = v + 1
            print(arr, "Vergleich:", v, "Swap:", s, "Vergleich von:", arr[j], arr[min] )
            if arr[j] < arr[min]:
                min = j
        if i != min:
            arr[i], arr[min] = arr[min], arr[i]
            s = s+1
            print("Swap of:" , arr[i], arr[min])
```

```python
A = [42, 17, 12, 15, 4, 11, 31, 14]
SelectionSort(A)
```

```python
Wir brauchen also 28 Vergleiche wenn wir Selection Sort verwenden!
```

```python
def InsertionSort(arr):
    v = 0   #Anzahl Vergleiche
    s = 0   #Anzahl swaps
    for i in range(1,len(arr)):
        for j in range(i,0,-1):
            v = v + 1
            if arr[j] < arr[j-1]:
                arr[j],arr[j-1] = arr[j-1],arr[j]
                s = s + 1
            print(arr, arr[j-1], arr[j], "Vergleich:", v, "Swap:", s)
```

```python
A = [42, 17, 12, 15, 4, 11, 31, 14]
InsertionSort(A)
```

```python
Wir brauchen hier also 22 Vergleiche, um den Array mit Insertion Sort zu Sortieren!

Warum stehen dann aber 28 Vergleiche hier im Output und nicht 22? 
    Tipp: Vergleiche die Python Implementierung aus der Übung
    mit der Java Implementierung aus der Vorlesung!
```

```python
#p.132
def quickSort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quickSort(arr, low, pi-1)
        quickSort(arr, pi+1, high)
    
def partition(arr, low, high):
    i = (low-1)
    pivot = arr[high]
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return(i+1)
```

```python
A = [42, 17, 12, 15, 4, 11, 31, 14]
quickSort(A,0,len(A)-1)
print(A)
```

```python
Heapsort: 
    
Garantiert: n*log(n) Laufzeit (nicht so auf Zufall abhängig wie bei QuickSort)

Allerdings bisschen komplizierter deshalb praktisch nicht schneller als QickSort
prinzipiell: Modifizierter Selection Sort: n mal das größte Element suchen,
        nur dass wir jetzt O(log(n)) brauchen im Schnitt für die Suche und nicht O(n)
        (denn: Tiefe von Binärbaum ist log2(n))

Heap = Binärbaum denn Arität = 2 (Binärbaum != Binärer Suchbaum (also necessary aber nicht sufficient)) 
     = mit allen Ebenen von Links Voll gefüllt ("Linksvollständig")
     = und mit Max Eigenschaft
    
Vorgehen: 
    
Heapify: Immer erst Max-Heap-Eigenschaft-Herstellen (von den Blättern an)
         Tausche Vorgängern mit Nachfolgern bis Heap Eigenschaft erfüllt
        
Sortierphase: Für alle Elemente:
– Tausche Wurzel (größtes Element) mit dem letztem Element.
– Stelle für alle Elemente kleiner 𝑥_𝑖 die Heapeigenschaft wieder her (Versickern)
                        
```
