```python
Maximum Subarray Problem: Formale Definiton:
    -Gegeben: Array mit positiven/negativen Zahlen
        
    -Gesucht: Den Unterarray finden, der die größte summe hat
```

```python
Zuerst machen betrachten wir die Naive Lösung, die zwei Loops macht, 
    das führt zum richtigen Ergebnis
    ist aber ineffizient - Frage: Welche Ordnung?
        
        
        
        
```

```python
idee: alle subarrays haben einen Anfangspunkt i
    
    -> wir gehen so alle möglichen subarrays durch indem wir von allen möglichen start-punken losgehen
    
    -> es wird in jeder Runde direkt verglichen ob der aktuelle subarray größer ist, als der größte bisherige
```

```python
def msp(a):
    resultat=0
    zwischensumme=0
    for i in range(len(a)):
        zwischensumme=0
        for j in range(i,len(a)):
            zwischensumme+=a[j]
            if zwischensumme>resultat:
                resultat=zwischensumme
            print(i,j,zwischensumme,resultat)
    return resultat
```

```python
a=[1,-3,4,-1,2,1,-5,4] #6 is the maxSubarray, as one can directly see via "eyeballing"
msp(a)
```

```python
basic idea: by induction: 

Falls  unter Null fällt,wäre es offensichtlich besser mit der Berechnung der Summe neu zu starten.
        
```

```python
def maxSubArrayKadane(nums):
    for i in range(1,len(nums)):
        if nums[i-1] > 0:
            nums[i] += nums[i-1]
        print(i,nums)
    return max(nums)
```

```python
a=[1,-3,4,-1,2,1,-5,4]
maxSubArrayKadane(a)
```

```python
-->O(n)
```

```python
#3-3 Eigenschaften von Bäumen

Wdh: Bäume (Kapitel 1):

Pfad: 
    
    
    
    
    
    Folge von Knoten, die durch Kanten direkt verbunden sind
    
    
    
    
    

Pfadlänge: 
    
    
    
    
    Anzahl der Kanten eines Pfades

    
    
    
    
    

    
    
Knotengrad: 
    
    
    
    Anzahl der unmittelbaren Nachfolger eines Knotens

    
    
    

    
Arität:
    
    
    
    
    
    
    Maximaler Knotengrad aller Knoten

    
      
        
    
    
    
    
Vollständiger Baum: 
    
    
    
    
    
    
    Hat JEDER Knoten den maximalen Grad, so spricht man von einem vollständigen Baum
    
    
    
    
    
    
    
Die Höhe eines Baums:
    
    
    
    
    ist die Pfadlänge des längsten Pfads von der Wurzel.
    
    
    
    
    

    
Binärbaum:
    
    
    
    
    Baum mit Arität 2 ist ein Binärbaum      
    
    
    
    

    
Für einen Binärbaum 𝑡, der Höhe ℎ gilt:

Max Zahl an Knoten:   
    
i. 𝑡 hat maximal 2^(ℎ+1) − 1 Knoten.






Min Zahl an Knoten:

ii. 𝑡 hat mindestens ℎ + 1 Knoten









Max Zahl an inneren Knoten:
iii. 𝑡 hat maximal 2^ℎ − 1 innere Knoten.







Max Zahl an Blättern:
iv. 𝑡 hat maximal 2^ℎ Blätter.
    
    

    
Arrayeinbettung von Bäumen:
    – Ebenen von der Wurzel an in das Array eintragen
    – Leere Positionen im Baum, im Array einfach freilassen
    
    


Bäume Durchlaufen

   4
 2   6
1,3,5,7


Tiefendurchlauf (DEPT FIRST):  Durchlaufe zu jedem Knoten rekursiv die Teilbäume von links nach rechts


– PREORDER/PRÄFIX/HAUPTREIHENFOLGE: (N–l–r) notiere erst einen Knoten, dann seine Teilbäume
    
    • Beispiel: 4 2 1 3 6 5 7 
        
– POSTORDER/POSTFIX/NEBENREIHENFOLGE: (l-r-N) notiere erst Teilbäume eines Knotens, dann ihn selbst
    
    • Beispiel: 1 3 2 5 7 6 4
    
– INORDER/INFIX/SYMMETRISCHE REIHENFOLGE: (l-N-r) notiere 1. Teilbaum, dann Knoten selbst, dann restliche Teilbäume
    
• Beispiel: 1 2 3 4 5 6 7
• Cool, weil es uns aus Binärem Suchbaum die Knoten sortiert zurück gibt
    
Breitendurchlauf (BREADTH FIRST):
– Knoten ebenenweise durchlaufen, von links nach rechts
• Beispiel: * - + 1 3 5 7
    
```

```python
from IPython.display import Math

Math(r'k_{max}=\sum_{i=0}^{i=h} n^i = \frac{n^{h+1}-1}{n-1}')

#Beweis siehe Tafel
```
